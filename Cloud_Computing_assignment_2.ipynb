```python
# Question 1 — Create a 20000 × 10000 Data File
import numpy as np

# Generate 20000 x 10000 matrix using different random generators
rows, cols = 20000, 10000

# Mix of different random distributions
data = np.zeros((rows, cols))
data[:5000, :]   = np.random.randint(0, 100, (5000, cols))          # integers
data[5000:10000,:] = np.random.uniform(0, 1, (5000, cols))          # uniform floats
data[10000:15000,:] = np.random.normal(50, 10, (5000, cols))        # normal dist
data[15000:, :]  = np.random.exponential(1.5, (5000, cols))         # exponential dist

# Save to file
np.save("data_matrix.npy", data)
print(f"Data file saved. Shape: {data.shape}")
print(f"File size: ~{data.nbytes / 1e6:.1f} MB")
```

```python
# Question 3 — Group Similar Elements into a Matrix
import numpy as np
from collections import defaultdict

data = np.random.randint(1, 6, (4, 5))  # values 1–5
print("Original Matrix:\n", data)

# Group positions by element value
groups = defaultdict(list)
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        groups[data[i, j]].append((i, j))

print("\nGrouped Elements:")
for value in sorted(groups):
    print(f"  Value {value}: positions {groups[value]}")

# Build a grouped matrix (each row = one group, padded with -1)
max_len = max(len(v) for v in groups.values())
grouped_matrix = np.full((len(groups), max_len), -1)
for idx, (val, positions) in enumerate(sorted(groups.items())):
    for col, pos in enumerate(positions):
        grouped_matrix[idx, col] = val  # store the value; positions stored above

print("\nGrouped Value Matrix (rows = unique values, padded with -1):\n", grouped_matrix)
```

```python
# Question 5 — n × n Square Matrix Where Sub-matrix Corner Sums Are Even
import numpy as np

def create_even_corner_matrix(n):
    """
    Creates an n x n matrix where for every 2x2 sub-matrix,
    the sum of opposite corner elements is even.
    Sum of diagonals (top-left + bottom-right) and (top-right + bottom-left) are both even.
    This holds when all elements are of the same parity (all even or all odd).
    """
    # Using all even numbers guarantees all sub-matrix corner sums are even
    matrix = np.array([[2 * (i * n + j + 1) for j in range(n)] for i in range(n)])
    return matrix

def verify_even_corners(matrix):
    n = matrix.shape[0]
    for i in range(n - 1):
        for j in range(n - 1):
            sub = matrix[i:i+2, j:j+2]
            s1 = int(sub[0,0]) + int(sub[1,1])  # main diagonal sum
            s2 = int(sub[0,1]) + int(sub[1,0])  # anti-diagonal sum
            if s1 % 2 != 0 or s2 % 2 != 0:
                return False, (i, j)
    return True, None

n = 4
mat = create_even_corner_matrix(n)
print(f"{n}x{n} Matrix:\n", mat)

valid, pos = verify_even_corners(mat)
print(f"\nAll 2x2 sub-matrix corner sums are even: {valid}")
```

```python
# Question 6 — Row-wise Element Addition in Tuple Matrix
# Tuple matrix: each element is a tuple of numbers; add element-wise across each row

tuple_matrix = [
    [(1, 2), (3, 4), (5, 6)],
    [(7, 8), (9, 1), (2, 3)],
    [(4, 5), (6, 7), (8, 9)],
]

print("Tuple Matrix:")
for row in tuple_matrix:
    print(" ", row)

print("\nRow-wise Element Addition (sum of all tuples in each row):")
for i, row in enumerate(tuple_matrix):
    # Add tuples element-wise across the row
    result = tuple(sum(tup[k] for tup in row) for k in range(len(row[0])))
    print(f"  Row {i}: {row} → Sum = {result}")
```
