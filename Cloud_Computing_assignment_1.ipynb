```python
# 1. Introduction to Google Colab Cloud
# Run this in a Colab cell to check the environment
import sys
print("Python version:", sys.version)
print("Platform:", sys.platform)

# 2. Interact with the Colab Environment
# Check available RAM and Disk
import psutil, shutil
ram = psutil.virtual_memory()
disk = shutil.disk_usage("/")
print(f"RAM  - Total: {ram.total/1e9:.2f} GB | Available: {ram.available/1e9:.2f} GB")
print(f"Disk - Total: {disk.total/1e9:.2f} GB | Free: {disk.free/1e9:.2f} GB")

# 3. Google Colab Computational Resources (check GPU/TPU)
import subprocess
try:
    result = subprocess.run(['nvidia-smi'], capture_output=True, text=True)
    print(result.stdout if result.returncode == 0 else "No GPU detected. Running on CPU.")
except FileNotFoundError:
    print("No GPU detected. Running on CPU.")

# 4. Python Environment in Google Colab
# List installed packages
import importlib.metadata
packages = sorted([f"{dist.name} {dist.version}" for dist in importlib.metadata.distributions()])
print("\nInstalled Packages (first 10):")
for p in packages[:10]:
    print(" ", p)
```
